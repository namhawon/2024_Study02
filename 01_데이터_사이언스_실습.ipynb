## **MBTI 테스트 문항 목록**

1. **주기적으로 새로운 친구를 사귄다.**
2. **복잡하고 참신한 아이디어는 단순하고 직관적인 아이디어보다 더 흥미진진합니다.**
3. **보통 사실에 기반한 주장보다 감정적으로 공감 가는 내용이 더 설득력 있다고 느낀다.**
4. **생활 공간과 업무 공간이 깨끗하고 정돈되어 있습니다.**
5. **압박감이 심한 환경에서도 평정심을 유지하는 편이다.**
6. **낯선 사람과 인적 네트워크를 형성하거나 나를 홍보하는 것은 매우 부담스러운 일입니다.**
7. **작업의 우선순위를 정하고 효과적으로 계획하여 마감일 전에 작업을 완료하는 경우가 많습니다.**
8. **숫자나 데이터보다는 사람들의 이야기와 감정이 마음에 더 와닿는다.**
9. **일정, 목록 같은 체계화 도구 사용을 좋아한다.**
10. **작은 실수로도 자신의 능력이나 지식을 의심할 때가 있다.**
11. **관심이 가는 사람에게 다가가서 대화를 시작하는 일을 편안하게 느낀다.**
12. **창작물의 다양한 해석에 대해 토론하는 일에는 큰 관심이 없다.**
13. **행동 계획을 결정할 때 사람들의 감정보다는 사실을 우선한다.**
14. **일정 없이 하루가 흘러가는 경우가 많습니다.**
15. **다른 사람에게 자신이 어떤 사람으로 보일지 걱정하는 경우는 드물다.**
16. **팀 기반 활동에 참여하는 것을 좋아한다.**
17. **새롭고 검증되지 않은 접근 방식을 실험하는 것을 즐깁니다.**
18. **완전한 솔직함보다 세심함을 우선시한다.**
19. **탐구할 새로운 경험, 지식 분야를 적극적으로 찾는다.**
20. **일이 잘못될까봐 자주 걱정하는 편이다.**
21. **단체 활동보다 혼자서 하는 취미나 활동을 더 좋아합니다.**
22. **소설 쓰는 일을 직업으로 삼은 나의 모습을 상상할 수 없다.**
23. **감정적 측면을 다소 무시하더라도 의사 결정의 효율성을 선호한다.**
24. **휴식을 취하기 전에 집안일을 먼저 하는 것을 더 좋아한다.**
25. **의견 차이가 있으면 다른 사람의 감정을 보호하기보다 자신의 주장을 입증하는 것을 우선시한다.**
26. **보통 사교 모임에서 다른 사람들이 자신을 먼저 소개할 때까지 기다립니다.**
27. **감정이 매우 빠르게 변할 때가 있다.**
28. **감정적인 논쟁에 쉽게 동요하지 않는다.**
29. **해야 할 일을 마지막까지 미룰 때가 많다.**
